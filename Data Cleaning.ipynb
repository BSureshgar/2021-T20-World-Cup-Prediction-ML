```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, glob
sns.set()
pd.set_option('display.max_rows', 2000)

teams = {'40': 'Afghanistan', '2' : 'Australia', '25' : 'Bangladesh', '1':'England','6':'India','29':'Ireland',
         '28':'Namibia','15':'Netherlands','5':'NewZealand','37':'Oman','7':'Pakistan','20':'PapuaNewGuinea',
         '30':'Scotland','3':'SouthAfrica','8':'SriLanka','4':'WestIndies'}
```

```python
def DataCleaner (team): 
    headers = ['Team', 'Result', 'Margin', 'BR', 'Toss', 'Bat','','Opposition', 'Ground', 'Date']
    raw_data = pd.read_csv('RawData/'+teams[str(team)]+'.csv', names=headers)
    
    working_data = raw_data.copy().iloc[1:]
    working_data = working_data.drop(['', 'BR', 'Toss', 'Bat','Margin'], axis = 1)
    working_data = working_data.reset_index() 
    working_data = working_data.drop(['index'], axis = 1)
    
    working_data.drop(working_data.index[working_data['Result'] == 'aban' ], inplace=True)
    working_data.drop(working_data.index[working_data['Result'] == 'n/r' ], inplace=True)
    
    working_data = working_data.dropna()
    
    working_data['Opposition'] = working_data['Opposition'].str.replace('v ','')
    
    working_data.to_csv(teams[str(team)] + "_cleaned.csv")
```

```python
for key in teams.keys():
    DataCleaner(key)
```

```python
path = "CleanedData/"
all_files = glob.glob(os.path.join(path, "*.csv"))
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged = pd.concat(df_from_each_file, ignore_index=True)
df_merged.to_csv("merged.csv")
fulldf = df_merged.copy()
```

```python
fulldf = fulldf.drop(['Unnamed: 0'], axis = 1)
```

```python
fulldf['Date']= fulldf['Date'].astype('datetime64[ns]')
fulldf = fulldf.drop_duplicates(subset=['Ground', 'Date'], keep='first')
fulldf.drop(fulldf.index[fulldf['Result'] == 'canc' ], inplace=True)
fulldf = fulldf.sort_values(by = 'Date', ascending = True)
fulldf = fulldf.reset_index(drop=True) 
fulldf 
```
