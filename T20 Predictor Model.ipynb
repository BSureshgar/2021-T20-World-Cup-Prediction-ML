```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#from xgboost import XGBClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
sns.set()
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
data = pd.read_csv("final_data.csv")
data = data.drop(['Unnamed: 0', 'Ground', 'Date'], axis = 1)
data
```

```python
final = pd.get_dummies(data, prefix=['Team1', 'Team2'], columns=['Team1', 'Team2'])
X = final.drop(['Winner'], axis=1)
y = final["Winner"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=28)
```

```python
y_train
```

```python
LogReg = LogisticRegression(random_state=0)
rf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=0)
clf = svm.SVC(kernel='linear')
#xgbmodel = XGBClassifier()

LogReg.fit(X_train, y_train) 
rf.fit(X_train, y_train) 
clf.fit(X_train, y_train) 
#xgbmodel.fit(X_train, y_train) 

score1 = LogReg.score(X_train, y_train)
score2 = LogReg.score(X_test, y_test)
score3 = rf.score(X_train, y_train)
score4 = rf.score(X_test, y_test)
score5 = clf.score(X_train, y_train)
score6 = clf.score(X_test, y_test)
#score7 = xgbmodel.score(X_train, y_train)
#score8 = xgbmodel.score(X_test, y_test)

print("LogReg Training set accuracy: ", '%.3f'%(score1))
print("LogReg Test set accuracy: ", '%.3f'%(score2))
print("RF Training set accuracy: ", '%.3f'%(score3))
print("RF Test set accuracy: ", '%.3f'%(score4))
print("SVM Training set accuracy: ", '%.3f'%(score5))
print("SVM Test set accuracy: ", '%.3f'%(score6))
```

```python
fixtures = pd.read_csv('Data/Fixtures.csv')
ranking = pd.read_csv('Data/ICCRankings.csv')
pred_set = []
fixtures.head()
```

```python
fixtures = fixtures.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 4','Unnamed: 7'], axis = 1)
headers = ['Team1', 'Team2', 'Venue', 'Result']
fixtures.columns = headers
fixtures = fixtures.copy().iloc[1:]
```

```python
# Create new columns with ranking position of each team
fixtures.insert(0, 'first_position', fixtures['Team1'].map(ranking.set_index('Team')['Position']))
fixtures.insert(1, 'second_position', fixtures['Team2'].map(ranking.set_index('Team')['Position']))

# We only need the group stage games, so we have to slice the dataset
fixtures = fixtures.iloc[:45, :]
fixtures
```

```python
# Loop to add teams to new prediction dataset based on the ranking position of each team
for index, row in fixtures.iterrows():
    if row['first_position'] < row['second_position']:
        pred_set.append({'Team1': row['Team1'], 'Team2': row['Team2'], 'winning_team': None})
    else:
        pred_set.append({'Team1': row['Team2'], 'Team2': row['Team1'], 'winning_team': None})
        
pred_set = pd.DataFrame(pred_set)
backup_pred_set = pred_set
pred_set
```

```python
# Get dummy variables and drop winning_team column
pred_set = pd.get_dummies(pred_set, prefix=['Team1', 'Team2'], columns=['Team1', 'Team2'])

# Add missing columns compared to the model's training dataset
missing_cols = set(final.columns) - set(pred_set.columns)
for c in missing_cols:
    pred_set[c] = 0
pred_set = pred_set[final.columns]


pred_set = pred_set.drop(['Winner'], axis=1)
pred_set_firstround = pred_set.iloc[:12]
```

```python
#group matches 
predictions_first = rf.predict(pred_set_firstround)
for i in range(pred_set_firstround.shape[0]):
    print(backup_pred_set.iloc[i, 1] + " vs " + backup_pred_set.iloc[i, 0])
    if predictions_first[i] == backup_pred_set.iloc[i, 1]:
        print("Winner: " + backup_pred_set.iloc[i, 1])
    
    else:
        print("Winner: " + backup_pred_set.iloc[i, 0])
    print("")
```
